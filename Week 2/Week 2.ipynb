```python
import pandas as pd
import numpy as np
```

```python
# URL of the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
iris_2d
```

```python
# Column names as they are not provided in the dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

# Load dataset into a Pandas DataFrame
df = pd.read_csv(url, header=None, names=column_names)

df.head()
```

```python
# Define two custom numpy arrays A and B
A = np.array([[1, 3, 6], [2, 5, 8]])
B = np.array([[4, 2, 5], [4, 8, 1]])
```

```python
# Stack A and B vertically and horizontally
vertical_stack = np.vstack((A, B))
horizontal_stack = np.hstack((A, B))

print("Vertical Stack:\n", vertical_stack)
print("Horizontal Stack:\n", horizontal_stack)
```

```python
# Common elements between A and B
common_elements = np.intersect1d(A, B)

print("Common Elements:\n", common_elements)
```

```python
# Extract numbers from A within a specific range (Here between 5 and 10)
filtered_A = A[(A >= 5) & (A <= 10)]
print("Filtered A (5-10):\n", filtered_A)
```

```python
# Filter rows where petal_length (3rd column) > 1.5 and sepal_length (1st column) < 5.0
filtered_iris = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)]

# Display results
print("Filtered Iris Data:\n", filtered_iris)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df
```

```python
# Select every 20th row starting from the 1st row (row index 0)
filtered_df = df.iloc[::20, :][['Manufacturer', 'Model', 'Type']]

# Display the filtered dataframe
print(filtered_df)
```

```python
# Replace missing values in Min.Price and Max.Price with their respective column mean
df.loc[:, 'Min.Price'] = df['Min.Price'].fillna(df['Min.Price'].mean())
df.loc[:, 'Max.Price'] = df['Max.Price'].fillna(df['Max.Price'].mean())

# Display the updated dataframe
print(df[['Min.Price', 'Max.Price']].head())  # Displaying only relevant columns
```
